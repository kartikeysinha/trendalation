## Example: all stocks

Take the base trend as S&P500. In a normal market environment with global economic factors, AAPL & S&P500 should be affected in a similar manner. However, when news comes out specific to AAPL, the trend should diverge.

This could be useful for portoflio managment when you want to see when your portfolio risk diverts from the model.

```python
s1 = snp500.loc['2019-01-01':, 'Open'].to_numpy()
s2 = aapl.loc['2019-01-01':, 'Open'].to_numpy()
s3 = tsla.loc['2019-01-01':, 'Open'].to_numpy()
s4 = msft.loc['2019-01-01':, 'Open'].to_numpy()

print(s1.shape, s2.shape, s3.shape, s4.shape)

# train just on snp500
print('Train on s1')
clf = PC()
clf.fit([s1])
print(clf.thresh)
print(clf.predict([s1, s2, s3, s4]))


print('Train on s2')
clf = PC()
clf.fit([s2])
print(clf.thresh)
print(clf.predict([s1, s2, s3, s4]))

print('Train on s3')
clf = PC()
clf.fit([s3])
print(clf.thresh)
print(clf.predict([s1, s2, s3, s4]))

print('Train on s4')
clf = PC()
clf.fit([s4])
print(clf.thresh)
print(clf.predict([s1, s2, s3, s4]))

print('Train on s1, s2')
clf = PC()
clf.fit([s1, s2])
print(clf.thresh)
print(clf.predict([s1, s2, s3, s4]))

print('Train on s1, s2, s4')
clf = PC()
clf.fit([s1, s2, s4])
print(clf.thresh)
print(clf.predict([s1, s2, s3, s4]))

print('Train on s1, s4')
clf = PC()
clf.fit([s1, s4])
print(clf.thresh)
print(clf.predict([s1, s2, s3, s4]))

print('Train on s1, s2, s3, s4')
clf = PC()
clf.fit([s1, s2, s3, s4])
print(clf.thresh)
print(clf.predict([s1, s2, s3, s4]))

print('Train on s2, s4')
clf = PC()
clf.fit([s2, s4])
print(clf.thresh)
print(clf.predict([s1, s2, s3, s4]))

print('Train on s3,')
clf = PC()
clf.fit([s3, s4])
print(clf.thresh)
print(clf.predict([s1, s2, s3, s4]))
```
